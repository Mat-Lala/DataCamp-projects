```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Loading CSV Files
lego_sets_df = pd.read_csv("datasets/lego_sets.csv")
parent_themes_df = pd.read_csv("datasets/parent_themes.csv")

# Inspecting the dataframes
print(lego_sets_df.head())
print(lego_sets_df.isna().sum())
```

```python

# There are some null values in our lego_sets that need to be removed
# Cleaning lego_sets_df DataFrame
lego_sets_df.dropna(subset = ["set_num"], inplace=True)
print(lego_sets_df.isna().sum())
```

```python

# To answer the questions we need to join our 2 Dataframes
lego_complete_df = lego_sets_df.join(parent_themes_df.set_index("name"), on="parent_theme")

# and drop all the non licensed sets
lego_sets_licensed_df = lego_complete_df[lego_complete_df["is_licensed"]==True]
print(lego_sets_licensed_df)
```

```python

# Calculating the percentage of Star Wars sets ever released
# 1st let's count how many sets there are per "theme", we'll use the groupby to do so
lego_sets_grouped = lego_sets_licensed_df.groupby("parent_theme")["id"].count()
print(lego_sets_grouped)
```

```python

# To calculate the percentage of Star Wars sets released, we just use our grouped by parent_theme dataframe
the_force = int(lego_sets_grouped[["Star Wars"]] / lego_sets_grouped.sum()*100)
print(the_force)
```

```python
# Let's use the groupby and pd.Series.mode to calculate which sets were the most popular per year
lego_sets_grouped_year = lego_sets_licensed_df.groupby("year")["parent_theme"].agg(pd.Series.mode) 
print(lego_sets_grouped_year)
```

```python
# We keep the only row that it not a Star Wars theme to answer the question:
# In which year was Star Wars not the most popular licensed theme
new_era = int(lego_sets_grouped_year[lego_sets_grouped_year != "Star Wars"].index[0])
print(new_era)
```
